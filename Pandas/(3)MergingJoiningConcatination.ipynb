```python
import numpy as np
import pandas as pd
```

```python
employees = pd.DataFrame({
    'Employee_ID':[1,2,3,4,5],
    'Name':['Raghav','Anjali','Pragya','Rahul','Aman'],
    'Department':['HR','IT','Finance','IT','HR']
})

Salaries = pd.DataFrame({
    'Employee_ID':[1,2,3,6,7],
    'Salary':[60000,80000,65000,70000,90000],
    'Bonus':[5000,10000,7000,8000,12000]
})
```

```python
employees
```

```python
Salaries
```

### Merjing Two DataFrame

```python
pd.merge(employees,Salaries,on='Employee_ID',how = 'inner')
```

```python
pd.merge(employees,Salaries,on='Employee_ID',how = 'outer')
```

```python
pd.merge(employees,Salaries,on='Employee_ID',how = 'left')
```

```python
pd.merge(employees,Salaries,on='Employee_ID',how = 'right')
```

### Concatination of 2 dataframes

```python
df1 = pd.DataFrame({
    'A':['A0','A1','A2'],
    'B':['B0','B1','B2'],
    'C':['C0','C1','C2']
})

df2 = pd.DataFrame({
    'A':['A3','A4','A5'],
    'B':['B3','B4','B5'],
    'C':['C3','C4','C5']
})
```

```python
df1
```

```python
df2
```

```python
pd.concat([df1,df2])
```

```python
pd.concat([df1,df2],axis = 1)
```

### Joining 2 data framess

```python
df1 = pd.DataFrame({
    'name': ['Alice', 'Bob', 'Charlie']
}, index=[1, 2, 3])

# Second DataFrame
df2 = pd.DataFrame({
    'score': [85, 90, 75]
}, index=[2, 3, 4])
```

```python
df1
```

```python
df2
```

```python
df1.join(df2,how = 'outer')
```

```python
df2.join(df1)
```
